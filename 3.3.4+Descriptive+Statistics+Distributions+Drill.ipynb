```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
plt.figure(figsize=(5,15))
 
plt.subplot(7,1,1)
power = np.random.power(5, 1000)
plt.hist(power)
#print('power mean:{}'.format(np.mean(power)))
#print('power stdev: {}'.format(np.std(power, ddof = 1)))
plt.axvline(np.mean(power), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(power) + np.std(power, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(power) - np.std(power, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.title("Power")
 
#The Mean is not a great representation of the central tendancy of this Power
#distribution because the distribution is skewed with the tail
#pointing to the left ("skewed to the left"). The standard deviation appears
#to capture far more than 68% as we see with the normal distribution.
 
plt.subplot(7,1,2)
gamma = np.random.gamma(4, 1, 1000)
plt.hist(gamma)
plt.axvline(np.mean(gamma), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(gamma) + np.std(gamma, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(gamma) - np.std(gamma, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(gamma), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Gamma")
 
#The Mean seems to be a decent representation of the central tendancy of this
#Gamma distribution. The standard deviation is less useful because the
#distribution slightly skews to the right.
 
plt.subplot(7,1,3)
geom = np.random.geometric(0.7, 1000)
plt.hist(geom)
plt.axvline(np.mean(geom), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(geom) + np.std(geom, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(geom) - np.std(geom, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(geom), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Geometric")
 
#The Mean is a good representation of the central tendancy of this
#Geometric distribution, but the standard deviation is nearly useless because
#the population is so densely clustered around the mean, and there are
#virtually no tails. Futhermore, given what the Geometric distribution
#represents, it is not even possible to have a value < 1, which is where one
#of the standard deviation bars lies.
 
plt.subplot(7,1,4)
chi = np.random.chisquare(3, 1000)
plt.hist(chi)
plt.axvline(np.mean(chi), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(chi) + np.std(chi, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(chi) - np.std(chi, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(chi), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Chi Square")
 
#The Mean seems to be a decent representation of the central tendancy of this
#Chi Square distribution (though not perfect because of skew), but the standard deviation is nearly useless because
#the population is so densely clustered around the mean, and there is only one
#tail.
 
plt.subplot(7,1,5)
tri = np.random.triangular(1, 5, 10, 1000)
plt.hist(tri)
plt.axvline(np.mean(tri), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(tri) + np.std(tri, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(tri) - np.std(tri, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(tri), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Triangle")
 
#The Mean and standard deviations are a great representations of the central
#tendencies and variance of the triangle distribution, depending on the parameters
#used in creating the random variabes. Since I used parameters that reflect
#a non-skewed distribution, the descriptive statistics are useful.
 
plt.subplot(7,1,6)
wb = np.random.weibull(3, 1000)
plt.hist(wb)
plt.axvline(np.mean(wb), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(wb) + np.std(wb, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(wb) - np.std(wb, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(wb), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Weibull")
 
#The Mean and standard deviations are a great representations of the central
#tendencies and variance of the Weibull distribution. This is because the
#population is most dense near the mean and has fairly even tails.
 
plt.subplot(7,1,7)
norm1 = np.random.normal(5, 0.5, 1000)
norm2 = np.random.normal(10, 1, 1000)
norm3 = norm1 + norm2
plt.hist(norm3)
plt.axvline(np.mean(norm3), color = 'red', linestyle = 'solid', linewidth = 2)
plt.axvline(np.mean(norm3) + np.std(norm3, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.axvline(np.mean(norm3) - np.std(norm3, ddof = 1), color = 'red', linestyle = 'dashed', linewidth = 2)
#plt.axvline(np.median(norm3), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title("Sum Normals")
 
#The Mean and standard deviations are a still great representations of the central
#tendencies and variance of the sum of two normal distributions.
#This is because the population is most dense near the mean and has fairly
#even tails (Also the sum of two normal distributions is still normal).
 
plt.tight_layout()
plt.show()
```
