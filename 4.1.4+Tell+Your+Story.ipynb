Section 1: Explore the landscape
I completed Harlan’s survey as if I had already completed the course, and it reported that I am a Data Creative with top skills in Math/OR. Given my OR background, I’d say that is a fairly accurate assessment. However, I found that I had very little or no experience with roughly half of the topics listed on the survey, which suggests I may have a long ways to go on reaching my goals... 

The Data Radar said that future me is a modeller with strong visualisation skills. The other skill categories were mid-range. Ideally, I would like to emerge from the program with strong data wrangling and programming skills too. It may be difficult to build strong programming skills through this course, though. In my mind, data scientists use programming languages but rely heavily upon statistical and analytical packages built by “traditional” programmers. I suppose the models and techniques I want to learn don’t require a rigorous programming background, after all. 

To me, a data scientist is someone that is capable of working with data from a variety of sources, including relational databases, APIs, web scraping, and more (CSV and JSON files, etc.). They also have the ability to generate insights through basic analysis and visualization (this is what a data analyst does, in my mind) and then create models to predict future behavior based (this is where data science branches off from data analyst). The data scientist is also more likely to follow current industry trends and new technologies to determine if they can improve upon their existing techniques.
